```python
import pandas as pd
import re
from bs4 import BeautifulSoup
import numpy as np
from collections import Counter
```

```python
okc = pd.read_csv('../Assets/A/profiles.csv')
```

```python
# Remove markdown

def essay_to_words( raw_essay ):

    # 1. Remove HTML
    text = BeautifulSoup(raw_essay, 'lxml').get_text() 
    #
    # 2. Remove Newline characters       
    text2 = text.replace('\n', ' ')
    return text2
```

Categories are:
books, movies, tv shows, music, food

```python
# Save counters for each category
books = Counter()
movies = Counter()
tv = Counter()
music = Counter()
authors = Counter()
# Save counters for all words that come before a colon
# There may be common variations on the above categories that have been overlooked
# Saving categories will help refine regex for search below
cats = Counter()
```

```python
# Takes regex search object and saves items from it to chosen list

def split_n_save(search_obj, cnt):

    for item in search_obj.split(' ',1)[1].rsplit(' ',1)[0].split(','):
        cnt[item] += 1 
```

### Search for movies, Tv Shows, Music, and food

```python
SA = okc.essay4[0]
```

```python
SA
```

```python
def find_lists(essay):
    
    # Skip empty essays
    # Remove markdown from the rest
    # Do any of these have capitol letters?

    try:
        SA = essay_to_words(essay)
    except:
        pass
    

    # Save lists for each category
    # All strings in Data Set A are lower case
    # Most in Data Set B contain lower and upper case
    
    try:
        book_list = re.search('[Bb]ooks?:(.*?)(:|$)|[Rreading:(.*?)(:|$)]', SA).group(0)
        split_n_save(book_list, books)
    except:
        pass
    
    try:
        author_list = re.search('[Aa]uthors:(.*?)(:|$)', SA).group(0)
        split_n_save(book_list, books)
    except:
        pass    
    
    try:
        movie_list = re.search('[Mm]ovies?:(.*?)(:|$)|[fF]ilms?:(.*?)(:|$)', SA).group(0)
        split_n_save(movie_list, movies)
    except:
        pass
    
    try:
        tv_list = re.search('([Tt][Vv]|[Ss]hows):(.*?)(:|$)', SA).group(0)
        split_n_save(tv_list, tv)
    except:
        pass
    try:
        music_list = re.search('([Mm]usic|[Bb]ands):(.*?)(:|$)', SA).group(0)
        split_n_save(music_list, music)
    except:
        pass
    
    try:
        food_list = re.search('[Ff]oods?:(.*?)(:|$)', SA).group(0)
        split_n_save(food_list, food)
    except:
        pass
    
    # Save other categories for later viewing
    try:
        for cat in re.findall('(\S*):', SA):
            cats[cat] += 1
    except:
        pass
```

```python
# Why does this keep printing 'None' for every single essay?
okc.essay4.apply(find_lists)
```
