### Create new df with top 10 features for men and top 10 features for women

```python
# Compare favorites for men vs. women

# Compute mean value of each of first 2000 columns for men
t0 = time()
mdf = faves[faves.sex=='m']
df = pd.DataFrame(mdf.iloc[:,:1-len(features)].mean(axis=0), columns = ['m'])
print time()-t0
```

```python
# Compute mean value of each of first 2000 columns for women
t0 = time()
fdf = faves[faves.sex == 'f']
df['f'] = fdf.iloc[:,:1-len(features)].mean(axis=0)
print time()-t0
```

```python
df['diff'] = df['m']-df['f']
df.sort_values('diff', inplace=True)
```

```python
# order top females words by 'f' column
df_f = df.head(10)
df_f = df_f.sort_values('f', ascending=False)

# order top male words by 'm' column
df_m = df.tail(10)
df_m = df_m.sort_values('m', ascending=True)

# Create new df with top 10 features for men and top 10 features for women
df_mf = df_f.append(df_m)
df_mf.to_csv('../Assets/A/faves_mf.csv')
```

### Explore tastes of heavy, social and non- drinkers

```python
# Compute mean value of each of columns for social drinks excluding features at end e.g. age, height, income
t0 = time()
ddf = faves[faves.drinks=='socially']
df = pd.DataFrame(ddf.iloc[:,:1-len(features)].mean(axis=0), columns = ['social'])
print time()-t0
```

```python
df.shape
```

```python
# Compute mean value of each of first 2000 columns for non-drinkers
t0 = time()
ddf = faves.query("drinks == 'rarely'|drinks == 'not at all'")
df['non-drinker'] = ddf.iloc[:,:1-len(features)].mean(axis=0)
print time()-t0

t0 = time()
ddf = faves.query("drinks == 'often'|drinks == 'very often'|drinks == 'desperately'")
df['heavy'] = ddf.iloc[:,:1-len(features)].mean(axis=0)
print time()-t0
```

```python
# Compute differences between social drinkers' values and other categories
df['nd_diff'] = df['non-drinker'] - df['social']
df['h_diff'] = df['heavy'] - df['social']

# Sort by top favorites for non-drinkers relative to social drinkers
nd_df = df.sort_values('nd_diff', ascending=False)

nd_df
```

```python
# Now sort by top favorites for heavy drinkers

h_df = df.sort_values('h_diff', ascending=False)

h_df
```

```python
# Save new df with top 8 faves from heavy and non-drinkers
d_faves = pd.concat([nd_df.head(8), h_df.head(8)], axis=0)
d_faves.to_csv('../Assets/A/d_faves.csv')
```

### Explore faves by age

```python
# Compare under 30 to over 30

def age_encoder(age):
    if age ==0:
        return np.nan
    elif age < 30:
        return 0
    else:
        return 1
    
faves.age = faves.age.apply(age_encoder)
```

```python
faves.age.value_counts()
```

```python

# Compute mean value of each of first 2000 columns for over 30
t0 = time()
odf = faves[faves.age==1]
df = pd.DataFrame(odf.iloc[:,:1-len(features)].mean(axis=0), columns=['>30'])
print time()-t0

# Compute mean value of each of first 2000 columns for under 30
ydf = faves[faves.age == 0]
df['<30'] = ydf.iloc[:,:1-len(features)].mean(axis=0)
print time()-t0
```

```python
df['diff'] = df['>30']-df['<30']
df.sort_values('diff', inplace=True)
```

```python
df
```

```python
# Create new df with top 10 features for <30 and top >30

faves_age = df.ix[df.index[:10]|df.index[-10:]]
faves_age.to_csv('../Assets/A/faves_age.csv')
```

### Explore faves of drug users vs. non-users

```python
def drug_encoder(drugs):
    if drugs == 'never':
        return 0
    elif drugs == 'sometimes' or drugs == 'often':
        return 1
    else:
        return np.nan
    
faves.drugs = faves.drugs.apply(drug_encoder)
```

```python
faves.drugs.value_counts()
```

```python
# Compute mean value of each of first 2000 columns for drug users
t0 = time()
udf = faves[faves.drugs==1]
df = pd.DataFrame(udf.iloc[:,:1-len(features)].mean(axis=0), columns = ['users'])
print time()-t0

# Compute mean value of each of first 2000 columns for non-users
ndf = faves[faves.drugs == 0]
df['non-users'] = ndf.iloc[:,:1-len(features)].mean(axis=0)
print time()-t0
```

```python
df['diff'] = df['users']-df['non-users']
df.sort_values('diff', inplace=True)
df
```

```python
# Create new df with top 10 features for <30 and top >30

faves_drugs = df.ix[df.index[:10]|df.index[-10:]]
faves_drugs.to_csv('../Assets/A/faves_drugs.csv')
```

### Explore faves by income

```python
# Divide income income into <50k, 50-100k, >100k

def income_encoder(income):
    if income == -1:
        return -1
    elif income <= 50000:
        return 0
    elif income <= 100000:
        return 1
    else:
        return 2
    
faves.income = faves.income.apply(income_encoder)
```

```python
faves.income.value_counts()
```

```python
# Compute mean value of each of first 2000 columns for lower income
t0 = time()
pdf = faves[faves.income==0]
df = pd.DataFrame(pdf.iloc[:,:1-len(features)].mean(axis=0), columns = ['<50k'])
print time()-t0

# Compute mean value of each of first 2000 columns for middle income
t0 = time()
mdf = faves[faves.income==1]
df['50-100k'] = mdf.iloc[:,:1-len(features)].mean(axis=0)
print time()-t0

# Compute mean value of each of first 2000 columns for highest income
t0 = time()
rdf = faves[faves.income==2]
df['>100k'] = rdf.iloc[:,:1-len(features)].mean(axis=0)
print time()-t0

# Compute differences between middle income values and others
df['r_diff'] = df['>100k'] - df['50-100k']
df['p_diff'] = df['<50k'] - df['50-100k']
```

```python
r_df = df.sort_values('r_diff', ascending=False)
r_df
```

```python
p_df = df.sort_values('p_diff', ascending=False)
p_df
```

```python

faves_income = pd.concat([r_df.head(8), p_df.head(8)], axis=0)
faves_income.to_csv('../Assets/A/faves_income.csv')
```

```python
faves_income
```

### Find faves of dog people vs cat people

```python
# calculate dummy columns for likes dogs and likes cats
# will be some overlap since not mutually exclusive

def mask_dogs(pets):
    # catches 'has dogs' or 'likes dogs'
    try:
        if pets.find('dogs') > -1:
            if pets.find('dislikes dogs') == -1:
                return 1
        else:
            return 0
    except:
        return 0
    
def mask_cats(pets):
    try:
        if pets.find('cats') > -1:
            if pets.find('dislikes cats') == -1:
                return 1
        else:
            return 0
    except:
        return 0

faves['dogs'] = faves.pets.apply(mask_dogs)
faves['cats'] = faves.pets.apply(mask_cats)
```

```python
# find mean scores for dog people

t0 = time()
ddf = faves[faves.dogs==1]
df = pd.DataFrame(ddf.iloc[:,:-1-len(features)].mean(axis=0), columns = ['dogs'])
print time()-t0

# find mean scores for cat people

cdf = faves[faves.cats == 1]
df['cats'] = cdf.iloc[:,:-1-len(features)].mean(axis=0)
print time()-t0
```

```python
df['diff'] = df['dogs'] - df['cats']
df.sort_values('diff', inplace=True)
```

```python
df
```

```python
# Save top 10 faves for dogs and cats to new df
faves_pets = df.ix[df.index[:10]|df.index[-10:]]
faves_pets.to_csv('../Assets/A/df_pets.csv')
```
